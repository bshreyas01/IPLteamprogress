```python
remteam=['Sunrisers Hyderabad','Gujarat Lions','Rising Pune Supergiant','Deccan Chargers','Kochi Tuskers Kerala',
 'Pune Warriors','Rising Pune Supergiants','Chennai Super Kings',
 'Rajasthan Royals']

for i in remteam:
    allteams.remove(i)
allteams
```

```python
appended_data = []
for teamname in allteams:
    ipl1=ipl[ipl['team1']==teamname]
    ipl3=ipl[ipl['team2']==teamname]
    teamm=ipl1.append(ipl3)
    teammwin=teamm[teamm['winner']==teamname]
    teammtotno=teamm.shape[0]
    teammwinno=teammwin.shape[0]
    seasonwinteamm=teammwin.groupby('season').count()
    seasontotteamm=teamm.groupby('season').count()
    seasongames=seasontotteamm['winner'].tolist()
    teammwinratio=seasonwinteamm['winner']/seasontotteamm['winner']
    appended_data.append(teammwinratio)
```

```python
appended_data
```

```python
customlines=[]
for i,j in zip(np.arange(len(allteams)),allteams):
    customlines.append(plt.plot(appended_data[i]))
    
```
